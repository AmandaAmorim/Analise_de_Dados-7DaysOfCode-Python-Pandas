```python
CDU_lista = ['Não classificado'] * len(df_completo) #criando lista com mesmo tamanho do dataframe
dicionario_cdu = {
    range(0, 90): 'Generalidades',
    range(100, 199): 'Filosofia e psicologia',
    range(200, 299): 'Religião',
    range(300, 399): 'Ciências sociais',
    range(400, 499): 'Classe vaga. Proviosoriamente não ocupada.',
    range(500, 599): 'Matemática e ciências naturais.',
    range(600, 699): 'Ciências aplicadas.',
    range(700, 799): 'Belas artes',
    range(800, 899): 'Linguagem.Língua.Linguística.',
    range(900, 999): 'Geografia.Biografia.História'
}

for i, CDU in enumerate(df_completo['localizacao']): #função enumerate retorna um índice e o valor para cada elemento da lista.
    for faixa, descricao in dicionario_cdu.items():
        if CDU in faixa:
            CDU_lista[i] = descricao
            break

df_completo['CDU_geral'] = CDU_lista
```
